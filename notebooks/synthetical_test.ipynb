```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_nodes = 100
dx = 0.01       # m
dt = 0.1        # s
D = 1.6e-5      # m/s (diffusion coefficient)
n_steps = 500

# Inertial air velocity parameters
tau = 1.0                              # seconds (response time for air inertia)
fan_center = int(n_nodes / 2)
fan_center = 0
fan_width = 10
fan_strength = 0.1                   # m/s (target velocity from fan)

# Initial air velocity
v = np.zeros(n_nodes)                 # actual air velocity (m/s)
v_target = fan_strength * np.exp(-((np.arange(n_nodes) - fan_center) / fan_width)**2)

# CO Concentration (mol/m)
C = np.zeros(n_nodes)
C_new = np.zeros_like(C)
source_concentration = 1.6

# For visualization
history = []
v_history = []

for t in range(n_steps):
    # Update air velocity with inertia
    dv_dt = (v_target - v) / tau
    v += dt * dv_dt

    # Source
    C[0] = source_concentration

    # Advection-Diffusion Update
    for i in range(1, n_nodes - 1):
        # Upwind advection with dynamic velocity
        if v[i] >= 0:
            adv_term = -v[i] * (C[i] - C[i-1]) / dx
        else:
            adv_term = -v[i] * (C[i+1] - C[i]) / dx

        diff_term = D * (C[i+1] - 2*C[i] + C[i-1]) / dx**2

        C_new[i] = C[i] + dt * (adv_term + diff_term)

    # Right boundary (no-flux)
    C_new[-1] = C[-1] + dt * (
        -v[-1] * (C[-1] - C[-2]) / dx +
        D * (C[-2] - C[-1]) / dx**2
    )

    # Update concentration
    C[1:] = C_new[1:]

    if t % 10 == 0:
        history.append(C.copy())
        v_history.append(v.copy())

# Plot CO profiles over time
positions = np.linspace(0, dx * n_nodes, n_nodes)
for i, snapshot in enumerate(history):
    plt.plot(positions, snapshot, label=f'{i*10*dt:.1f} s')

plt.xlabel('Position (m)')
plt.ylabel('CO Concentration (mol/m)')
plt.title('CO Diffusion with Inertial Fan Advection')
plt.legend()
plt.grid(True)
plt.show()

# Plot velocity field over time
for i, v_snap in enumerate(v_history):
    plt.plot(positions, v_snap, label=f'{i*10*dt:.1f} s')

plt.xlabel('Position (m)')
plt.ylabel('Air Velocity (m/s)')
plt.title('Inertial Response of Air Velocity (Fan Influence)')
plt.legend()
plt.grid(True)
plt.show()
```
